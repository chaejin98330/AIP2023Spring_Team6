# Feature map visualization

## settings

```python
import numpy as np
import torch
import matplotlib.pyplot as plt
from PIL import Image
import timm
```

```python
video_id = 'zbsW6FeS0n0'
path = f"./medium_15287/{video_id}.jpg"
device = "cuda:0" if torch.cuda.is_available() else "cpu"
device
```

## Feature map

```python
import matplotlib.pyplot as plt

image = Image.open(path)
image = torch.tensor(np.array(image), dtype=torch.float32, device=device).permute(2, 0, 1).unsqueeze(dim=0)
# feature_map shape: (320, 6, 10)
model = timm.create_model('efficientnet_b1_pruned', features_only =True, pretrained=True)
model.eval()
feature_map = model(image)[-1].squeeze() # (320,6,10)
np_fmap = feature_map.detach().numpy()

# channel to 2D image
channel_images = [np_fmap[channel_idx, :, :] for channel_idx in range(np_fmap.shape[0])]

# each channel visualization
fig, axes = plt.subplots(nrows=8, ncols=8, figsize=(100, 100))
for idx, ax in enumerate(axes.flat):
    ax.imshow(channel_images[idx], cmap='gray')
    ax.axis('off')

plt.show()
```

## Transform feature map

```python
import matplotlib.pyplot as plt

image = Image.open(f'./medium_15287/{video_id}.jpg')
# feature_map shape: (320, 6, 10)
model = timm.create_model('efficientnet_b1_pruned', features_only =True, pretrained=True)
model.eval()
data_cfg = timm.data.resolve_data_config(model.pretrained_cfg)
transform = timm.data.create_transform(**data_cfg)
image = transform(image).unsqueeze(dim=0)

feature_map = model(image)[-1].squeeze() # (320,6,10)
np_fmap = feature_map.detach().numpy()

# channel to 2D image
channel_images = [np_fmap[channel_idx, :, :] for channel_idx in range(np_fmap.shape[0])]

# each channel visualization
fig, axes = plt.subplots(nrows=8, ncols=8, figsize=(100, 100))
for idx, ax in enumerate(axes.flat):
    ax.imshow(channel_images[idx], cmap='gray')
    ax.axis('off')

plt.show()
```
